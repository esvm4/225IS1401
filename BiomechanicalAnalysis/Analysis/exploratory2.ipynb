# Analysis of joint angle variations across conditions, subjects, and replications
1. Read Data 
2. Data Description 
3. Analyze the difference between conditions
4. Analyze the difference between subject
5. Analyze the difference between replication

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('./gait.csv')
df.head()
```

```python
mean_angle = df.groupby('subject')['angle'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(mean_angle['subject'], mean_angle['angle'])
plt.xlabel('Subject')
plt.ylabel('Mean Angle')
plt.title('Comparison of Joint Angle by Subject')
plt.show()
```

```python
mean_angle = df.groupby('condition')['angle'].mean().reset_index()

plt.figure(figsize=(5, 3))
plt.bar(mean_angle['condition'], mean_angle['angle'])
plt.xlabel('Condition')
plt.ylabel('Mean Angle')
plt.title('Comparison of Joint Angle by Condition')
plt.show()
```

```python
mean_angle = df.groupby('replication')['angle'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(mean_angle['replication'], mean_angle['angle'])
plt.xlabel('Replication')
plt.ylabel('Mean Angle')
plt.title('Comparison of Joint Angle by Replicationt')
plt.show()
```

```python
sns.histplot(data=df, x='angle', hue='condition', bins = 40, kde=True);
```

```python
mean_angle = ['-90° - 45°', '-45° - 0', '0 - 45°', '45° - 90°']
quartile_data = pd.qcut(df['angle'], 4, labels=mean_angle)

sns.countplot(x=quartile_data, hue=df['condition'], palette='Blues')
```
